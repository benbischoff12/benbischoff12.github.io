```python
#Author 1
import os
import requests
import re
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer
)
os.environ["WANDB_DISABLED"] = "false"
# ==== 1. DOWNLOAD AND CLEAN THE TEXT ====

book_url = "https://www.gutenberg.org/files/11/11-0.txt"
local_dir = "saved_files"
local_path = os.path.join(local_dir, "11-0.txt")

os.makedirs(local_dir, exist_ok=True)

# Download Alice in Wonderland
response = requests.get(book_url)
with open(local_path, "w", encoding="utf-8") as f:
    f.write(response.text)

# Read and extract core text
with open(local_path, "r", encoding="utf-8") as f:
    book_text = f.read()

start = book_text.find("CHAPTER I.")
end = book_text.find("*** END OF THE PROJECT GUTENBERG EBOOK")
book_text = book_text[start:end].strip()

# Basic cleanup
def preprocess_text(text):
    text = text.replace("\r\n", " ").replace("\n", " ")
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"\s([?.!,;:](?:\s|$))", r"\1", text)
    return text.strip()

book_text = preprocess_text(book_text)

# Split into paragraphs
paragraphs = [p.strip() for p in book_text.split(". ") if len(p.strip()) > 50]
dataset = Dataset.from_dict({"text": paragraphs})

# ==== 2. TOKENIZE THE DATASET ====

tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # GPT-2 needs a pad token

def tokenize_function(example):
    tokenized = tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=512
    )
    tokenized["labels"] = tokenized["input_ids"].copy()  # ✅ this line is key
    return tokenized

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# ==== 3. LOAD THE MODEL ====

model = AutoModelForCausalLM.from_pretrained("gpt2")
model.resize_token_embeddings(len(tokenizer))  # in case vocab was changed

# ==== 4. SET TRAINING CONFIG ====

training_args = TrainingArguments(
    output_dir="./gpt2-alice",
    per_device_train_batch_size=2,
    num_train_epochs=20,
    save_steps=500,
    save_total_limit=2,
    logging_steps=100,
    fp16=torch.cuda.is_available(),  # Use mixed precision if on GPU
    overwrite_output_dir=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer
)

# ==== 5. TRAIN THE MODEL ====

trainer.train()

# ==== 6. SAVE THE MODEL ====

model.save_pretrained("gpt2-alice")
tokenizer.save_pretrained("gpt2-alice")

print("✅ Training complete. Model saved to ./gpt2-alice")
```
